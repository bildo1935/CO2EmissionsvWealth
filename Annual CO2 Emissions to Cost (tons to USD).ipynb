```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def emissions2cost(tons, pertontax):
    cost = pertontax * tons
    return cost

#pertontax derived from Report on the High-Level Commission of Carbon Pricing by Carbon Pricing Leadership Coalition
#Report can be downloaded from (https://www.carbonpricingleadership.org/report-of-the-highlevel-commission-on-carbon-prices)

df = pd.read_csv("Global Annual CO2 Emissions.csv")
df1 = pd.read_csv("Cleaned Global Population Data.csv")
df['Annual CO2 emissions'] = df['Annual CO2 emissions']/df1['Population'] #Division by population size gets per capita data for cost

low_cost_est = [] #low_cost_est represents optimistic estimates of the per capita cost of carbon emissions according to lower limit of suggested per-ton-tax of 45 USD/ton
high_cost_est =[] #low_cost_est represents pessimistic estimates of the per capitacost of carbon emissions according to upper limit of suggested per-ton-tax of 90 USD/ton

for cell in df['Annual CO2 emissions']:
    low = emissions2cost(cell, 45)
    high = emissions2cost(cell, 90)
    low_cost_est.append(low)
    high_cost_est.append(high)
    
df['High Cost Estimates (USD)'] = high_cost_est
df['Low Cost Estimates (USD)'] = low_cost_est
```

```python
df.plot(x='Year')
```

```python
df.to_csv("Annual CO2 Emissions to Cost (tons to USD).csv")
```
